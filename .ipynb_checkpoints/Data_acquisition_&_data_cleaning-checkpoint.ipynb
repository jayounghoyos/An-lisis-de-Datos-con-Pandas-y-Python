### Limpieza de los ***datos***

```python
columns_to_drop = ["PassengerId", "Name", "Ticket", "Cabin", "Embarked", "Fare"]

# errors='ignore' prevents KeyError if columns were already dropped (e.g. re-running this cell)
dataset = dataset.drop(columns=columns_to_drop, errors='ignore')

print("Columns after dropping:", dataset.columns.tolist())
print("Shape:", dataset.shape)
dataset.head()
```

```python
# fillna is already safe to re-run (filling non-NaN values has no effect)
dataset = dataset.fillna({"Age": dataset["Age"].mean()})

print("Remaining nulls:")
print(dataset.isnull().sum())
print("\nCleaned dataset:")
dataset.head()
```

```python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")

df = df.drop("Ticket", axis = 1)
df = df.drop("PassengerId", axis = 1)
df = df.drop("Embarked", axis = 1)
df = df.drop("Fare", axis = 1)
df = df.drop("Name", axis = 1)
df = df.drop("Cabin", axis = 1)

df = df.fillna({"Age" : df["Age"].mean()})

print(df)
```
