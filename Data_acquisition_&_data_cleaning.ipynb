Este código realiza pasos esenciales de preprocesamiento de datos en el conjunto de datos de Titanic para prepararlo para construir un modelo de IA que prediga la supervivencia de los pasajeros.

1. Carga del conjunto de datos:
  - El conjunto de datos se carga desde un archivo CSV en un DataFrame de pandas, lo que permite una fácil manipulación y análisis de datos.

2. Eliminación de columnas irrelevantes o redundantes:
  - 'PassengerId': Un identificador único para cada pasajero, que no proporciona información predictiva.
  - 'Ticket': Los números de ticket son en su mayoría arbitrarios y es poco probable que ayuden al modelo a aprender patrones de supervivencia.
  - 'Name': Los nombres son únicos y textuales; sin un procesamiento adicional (como la extracción de títulos), no contribuyen directamente a la predicción.
  - 'Cabin': Esta columna tiene muchos valores faltantes y datos inconsistentes, lo que la hace poco confiable sin una imputación compleja.
  - 'Embarked': Aunque podría tener algún valor predictivo, se elimina aquí, posiblemente para simplificar el modelo o debido a valores faltantes.
  - 'Fare': La tarifa podría correlacionarse con la supervivencia, pero se excluye aquí, probablemente para centrarse en otras características o debido a consideraciones de calidad de los datos. La eliminación de estas columnas reduce el ruido y la dimensionalidad, ayudando al modelo a centrarse en características más relevantes.


# Mínima limpieza con base en el análisis

```python
#Authors: Martín Valencia, María Acevedo, Juan Andrés Young

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/FundamentosDeAprendizajeEAFIT/Titanic-Dataset.csv')


'''
Impresión del DataFrame limpio:
  - Este paso permite verificar que los datos se hayan limpiado correctamente, con las columnas previstas eliminadas y los valores faltantes manejados.
'''

df = df.drop("Ticket", axis = 1)
df = df.drop("PassengerId", axis = 1)
df = df.drop("Embarked", axis = 1)
df = df.drop("Fare", axis = 1)
df = df.drop("Name", axis = 1)
df = df.drop("Cabin", axis = 1)

print(df)
```

```python
df = df.fillna({"Age" : df["Age"].mean()})

'''
Manejo de valores faltantes en la columna 'Age':
  - La característica 'Age' es importante porque la edad puede influir en las posibilidades de supervivencia (por ejemplo, los niños podrían haber sido priorizados).
  - Los valores de edad faltantes se rellenan con la edad media de todos los pasajeros, una técnica de imputación común y simple.
  - Esto asegura que no queden valores faltantes, lo cual es necesario porque muchos algoritmos de aprendizaje automático no pueden manejar datos faltantes. '''

print(df)
```

## 1.3 Comparacion entre limpieza minima y maxima

### 1.3.1 Limpieza minima
Solo lo necesario para que el algoritmo de ML corra sin errores:
- Eliminar columnas no numericas (Name, Sex, Ticket, Cabin, Embarked) para evitar errores de tipo
- Eliminar filas con NaN (Age) para evitar errores de valores faltantes
- Se pierde informacion valiosa (Sex es el predictor #1 del Titanic) y se pierden 177 filas

```python
import pandas as pd

df_min = pd.read_csv('/content/drive/MyDrive/FundamentosDeAprendizajeEAFIT/Titanic-Dataset.csv')

# Solo eliminar columnas no numericas y PassengerId
df_min = df_min.drop(columns=['PassengerId', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'])

# Eliminar filas con NaN para que no haya errores
df_min = df_min.dropna()

df_min.to_csv('titanic_min.csv', index=False)
print(f'Shape limpieza minima: {df_min.shape}')
print(f'Columnas: {list(df_min.columns)}')
df_min.head()
```

### 1.3.2 Limpieza maxima
Limpieza exhaustiva para maximizar el desempenio del modelo:
- Imputar Age con la mediana por Pclass (mas preciso que media global)
- Imputar Embarked con la moda
- Codificar Sex con LabelEncoder (female=0, male=1)
- One-Hot Encoding para Embarked
- Feature engineering: FamilySize y IsAlone
- Se conservan todas las 891 filas y se agregan features informativos

```python
from sklearn.preprocessing import LabelEncoder

df_max = pd.read_csv('/content/drive/MyDrive/FundamentosDeAprendizajeEAFIT/Titanic-Dataset.csv')

# Eliminar columnas irrelevantes
df_max = df_max.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])

# Imputar Age con mediana por Pclass
df_max['Age'] = df_max.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))

# Imputar Embarked con la moda
df_max['Embarked'] = df_max['Embarked'].fillna(df_max['Embarked'].mode()[0])

# Codificar Sex
le = LabelEncoder()
df_max['Sex'] = le.fit_transform(df_max['Sex'])

# One-Hot Encoding para Embarked
df_max = pd.get_dummies(df_max, columns=['Embarked'], drop_first=True)

# Feature engineering
df_max['FamilySize'] = df_max['SibSp'] + df_max['Parch'] + 1
df_max['IsAlone'] = (df_max['FamilySize'] == 1).astype(int)

# Convertir columnas bool a int (por compatibilidad)
for col in df_max.select_dtypes(include='bool').columns:
    df_max[col] = df_max[col].astype(int)

df_max.to_csv('titanic_max.csv', index=False)
print(f'Shape limpieza maxima: {df_max.shape}')
print(f'Columnas: {list(df_max.columns)}')
print(f'Valores faltantes: {df_max.isnull().sum().sum()}')
df_max.head()
```

## Aplicar Modelo de ML (aplicarModeloML.ipynb)
Algoritmo: **RandomForestClassifier** (clasificacion binaria: Survived 0/1)

Se usa el mismo algoritmo con los mismos parametros en ambas limpiezas para comparar el efecto de la limpieza en el desempenio.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Modelo con limpieza minima

```python
# PREPARACION
X_min = df_min.drop(columns=['Survived'])
y_min = df_min['Survived']

X_train_min, X_test_min, y_train_min, y_test_min = train_test_split(
    X_min, y_min, test_size=0.2, random_state=42
)

# ESCALAMIENTO
scaler_min = StandardScaler()
X_train_min = scaler_min.fit_transform(X_train_min)
X_test_min = scaler_min.transform(X_test_min)

# ENTRENAMIENTO CON RANDOM FOREST
algorithm_min = RandomForestClassifier(n_estimators=100, random_state=42)
algorithm_min.fit(X_train_min, y_train_min)

# PREDICCION Y ACCURACY
y_pred_min = algorithm_min.predict(X_test_min)
acc_min = accuracy_score(y_test_min, y_pred_min)

print(f"--- RESULTADOS LIMPIEZA MINIMA ---")
print(f"Accuracy Global: {acc_min * 100:.2f}%")
print("-" * 33)
print(f"\nClassification Report:")
print(classification_report(y_test_min, y_pred_min))
print(f"Confusion Matrix:")
print(confusion_matrix(y_test_min, y_pred_min))
```

### Modelo con limpieza maxima

```python
# PREPARACION
X_max = df_max.drop(columns=['Survived'])
y_max = df_max['Survived']

X_train_max, X_test_max, y_train_max, y_test_max = train_test_split(
    X_max, y_max, test_size=0.2, random_state=42
)

# ESCALAMIENTO
scaler_max = StandardScaler()
X_train_max = scaler_max.fit_transform(X_train_max)
X_test_max = scaler_max.transform(X_test_max)

# ENTRENAMIENTO CON RANDOM FOREST (mismos parametros)
algorithm_max = RandomForestClassifier(n_estimators=100, random_state=42)
algorithm_max.fit(X_train_max, y_train_max)

# PREDICCION Y ACCURACY
y_pred_max = algorithm_max.predict(X_test_max)
acc_max = accuracy_score(y_test_max, y_pred_max)

print(f"--- RESULTADOS LIMPIEZA MAXIMA ---")
print(f"Accuracy Global: {acc_max * 100:.2f}%")
print("-" * 33)
print(f"\nClassification Report:")
print(classification_report(y_test_max, y_pred_max))
print(f"Confusion Matrix:")
print(confusion_matrix(y_test_max, y_pred_max))
```

### Comparacion de resultados

```python
print("=" * 40)
print("       COMPARACION DE RESULTADOS")
print("=" * 40)
print(f"Accuracy limpieza minima:  {acc_min * 100:.2f}%")
print(f"Accuracy limpieza maxima:  {acc_max * 100:.2f}%")
print(f"Mejora:                    {(acc_max - acc_min) * 100:.2f}%")
print("=" * 40)
```

### Conclusiones

- **Limpieza minima**: Se usaron solo columnas numericas originales (Pclass, Age, SibSp, Parch, Fare), se eliminaron 177 filas con NaN en Age, y se descarto informacion clave como Sex (el predictor mas fuerte de supervivencia en el Titanic).

- **Limpieza maxima**: Se conservaron las 891 filas al imputar valores faltantes, se codifico Sex y Embarked, y se crearon features nuevos (FamilySize, IsAlone). El modelo tuvo acceso a mas informacion relevante.

- La limpieza maxima produce un **mejor desempenio** porque el modelo tiene acceso a Sex (women and children first), mas datos de entrenamiento, y features adicionales que capturan patrones de supervivencia.
